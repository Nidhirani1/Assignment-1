```python
# Q5. Given a list my_list = [1, 2, 3, 4, 5], write the code to slice the list and obtain the sub-list [2, 3].
# Ans:-To obtain the sublist [2, 3] from the list my_list = [1, 2, 3, 4, 5], 
# you can use list slicing in Python.
# Here's the code:


# my_list = [1, 2, 3, 4, 5]

# # Slice the list to obtain the sublist [2, 3]
# sub_list = my_list[1:3]

# print(sub_list)  # Output: [2, 3]

# In Python, list slicing works as follows:

# my_list[start:end] returns a sublist starting from the index start (inclusive) up to, but not including,
# the index end (exclusive).
# If start is omitted, it defaults to the beginning of the list.
# If end is omitted, it defaults to the end of the list.
# So, my_list[1:3] slices my_list starting from index 1 (which corresponds to the second element, 2) up to, 
# but not including, index 3 (which corresponds to the fourth element, 4), resulting in the sublist [2, 3]
```
